## Data Understanding
### Pengertian Data Understanding
Data Understanding adalah tahap dalam proses Data Mining yang bertujuan untuk memahami karakteristik, struktur, dan kualitas data yang akan digunakan dalam analisis.

Setelah pada tahap sebelumnya kita memahami masalah bisnis, maka pada tahap ini kita mulai memahami:

    Data apa yang tersedia dan apakah data tersebut mampu menjawab masalah bisnis?
    
Tahap ini menjadi jembatan antara kebutuhan bisnis dan proses teknis analisis data.

### Tujuan Data Understanding
Tujuan utama tahap ini adalah:
1. Mengumpulkan data yang relevan
1. Mengenali struktur dan tipe data
1. Menilai kualitas data
1. Mengidentifikasi pola awal
1. Menemukan potensi masalah dalam data

Tahap ini membantu memastikan bahwa data benar-benar sesuai dengan tujuan proyek.

### Mengapa Data Understanding Penting?
Tanpa memahami data dengan baik:
* Model bisa dibangun dari data yang salah
* Analisis menjadi tidak akurat
* Kesimpulan bisa menyesatkan

Contoh:
Jika ingin memprediksi churn pelanggan tetapi data tidak memiliki informasi riwayat transaksi, maka prediksi menjadi kurang akurat.

### Tahapan dalam Data Understanding
1. Pengumpulan Data (Data Collection)
Data dapat berasal dari berbagai sumber, seperti:
    * Database internal perusahaan
    * File CSV atau Excel
    * API
    * Data publik
    * Sistem transaksi
    
    Pada tahap ini, penting untuk memastikan bahwa data yang dikumpulkan relevan dengan tujuan bisnis.
1. Deskripsi Data (Describing Data)
Langkah berikutnya adalah memahami struktur data.

    Hal-hal yang diperiksa:
    * Jumlah baris dan kolom
    * Nama variabel
    * Tipe data (numerik, kategorikal, teks, tanggal)
    * Rentang nilai setiap variabel

    Contoh:
    * Kolom "Umur" bertipe numerik
    * Kolom "Jenis Kelamin" bertipe kategorikal
    * Kolom "Tanggal Transaksi" bertipe datetime

1. Exploratory Data Analysis (EDA)
Exploratory Data Analysis (EDA) bertujuan untuk menemukan pola awal dalam data.

    Kegiatan yang dilakukan:
    * Menghitung statistik deskriptif (mean, median, standar deviasi)
    * Melihat distribusi data (histogram)
    * Menganalisis hubungan antar variabel (korelasi)
    * Mengidentifikasi outlier

    EDA membantu memahami karakteristik data sebelum masuk ke tahap pembersihan dan pemodelan.
1. Evaluasi Kualitas Data
Data dunia nyata sering memiliki masalah, seperti:
    * Missing values (nilai kosong)
    * Duplikasi data
    * Kesalahan input
    * Outlier ekstrem
    * Inkonsistensi format

    Pada tahap ini, semua potensi masalah tersebut diidentifikasi untuk diperbaiki pada tahap Data Preparation.

### Contoh Sederhana
Misalkan kita memiliki data pelanggan dengan kolom:
* ID Pelanggan

* Pendapatan

Pada tahap Data Understanding, kita:

Dari sini kita mulai mendapatkan insight awal.

### Hubungan dengan Tahap Lain
Data Understanding berada di antara:

Business Understanding $\rightarrow$ Data Understanding $\rightarrow$ Data Preparation

Jika pada tahap ini ditemukan bahwa data tidak lengkap atau tidak relevan, maka kita mungkin perlu:

Proses ini bersifat iteratif.

#### C. Statistik Deskriptif menggunakan Python
Kode berikut digunakan untuk menghitung statistik deskriptif pada dataset IRIS, yang bertujuan untuk memberikan gambaran umum mengenai karakteristik data, seperti jumlah data, rata-rata, nilai minimum, maksimum, kuartil, standar deviasi, variansi, skewness, dan modus.

```python
df_numeric = df.select_dtypes(include=['int64','float64'])

for col in df_numeric.columns:
    print("Statistik untuk", col,)
    print("Jumlah Data      :", df[col].count())
    print("Rata-rata        :", df[col].mean())
    print("Nilai Minimum    :", df[col].min())
    print("Q1               :", df[col].quantile(0.25))
    print("Median (Q2)      :", df[col].quantile(0.5))
    print("Q3               :", df[col].quantile(0.75))
    print("Nilai Maximum    :", df[col].max())
    print("Kemencengan      :", round(df[col].skew(), 6))
    mode = df[col].mode()
    print("Modus            :", mode.iloc[0] if not mode.empty else "Tidak ada modus")
    print("Standar Deviasi  :", round(df[col].std(), 6))
    print("Variansi         :", round(df[col].var(), 6))
    print("\n")
```

Penjelasan Kode:
* df_numeric = df.select_dtypes(include=['int64','float64']) Digunakan untuk mengambil hanya kolom bertipe numerik agar kolom kategorikal tidak ikut dihitung statistiknya.
* for col in df_numeric.columns: Digunakan untuk melakukan perulangan pada setiap kolom numerik.
* df[col].count() Menghitung jumlah data (tidak termasuk nilai kosong).
* df[col].mean() Menghitung nilai rata-rata.
* df[col].min() dan df[col].max() Menentukan nilai minimum dan maksimum.
* df[col].quantile(0.25 / 0.5 / 0.75) Menghitung kuartil:
    * Q1 (25%)
    * Median / Q2 (50%)
    * Q3 (75%)
* df[col].skew() Menghitung kemencengan (skewness) distribusi data.
* df[col].mode() Menghitung nilai yang paling sering muncul (modus).
* mode.iloc[0] if not mode.empty digunakan untuk mengambil nilai modus pertama jika ada.
* df[col].std() Menghitung standar deviasi.
* df[col].var() Menghitung variansi.
* round(..., 6) Digunakan untuk membulatkan hasil perhitungan menjadi 6 angka di belakang koma.

**Cek Outlier**
  
Outlier adalah data yang memiliki nilai ekstrem dan berbeda jauh dari mayoritas data lainnya.

Pada dataset Iris, outlier bisa muncul jika ada nilai ukuran bunga (misalnya panjang sepal atau petal) yang terlalu besar atau terlalu kecil dibandingkan data lainnya.

**Cara Menentukan Outlier pada Iris**

Biasanya digunakan metode IQR (Interquartile Range): $$IQR = Q_3 - Q_1$$

Batas bawah: $$\text{Batas Bawah} = Q_1 - 1.5 \times IQR$$

Batas atas: $$\text{Batas Atas} = Q_3 + 1.5 \times IQR$$

**Apakah Dataset Iris Memiliki Outlier?**
Secara umum:
  * Dataset Iris relatif bersih
  * Tidak banyak outlier ekstrem
  * Namun pada beberapa fitur seperti sepal width, kadang muncul beberapa nilai yang terdeteksi sebagai outlier berdasarkan metode IQR

Tetapi outlier tersebut masih termasuk variasi alami ukuran bunga, bukan kesalahan data.

**Kesimpulan Outlier:**

Outlier pada dataset Iris adalah nilai ukuran bunga yang berada jauh di luar rentang mayoritas data. Berdasarkan metode IQR, dataset Iris umumnya tidak memiliki outlier ekstrem yang signifikan, sehingga dataset ini tergolong bersih dan cocok digunakan untuk analisis atau machine learning.

```python
outlier_summary = []
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    jumlah_outlier = df[(df[col] < lower) | (df[col] > upper)].shape[0]
    outlier_summary.append([col, jumlah_outlier])
outlier_table = pd.DataFrame(outlier_summary, columns=["Kolom", "Jumlah Outlier"])
outlier_table
```

Penjelasan Kode:
* outlier_summary = [] digunakan untuk membuat list kosong yang akan menyimpan ringkasan jumlah outlier setiap kolom numerik.
* for col in numeric_cols: digunakan untuk melakukan perulangan pada setiap kolom numerik yang terdapat dalam variabel numeric_cols.
* Q1 = df[col].quantile(0.25) digunakan untuk menghitung kuartil pertama (25%) dari kolom tersebut.
* Q3 = df[col].quantile(0.75) digunakan untuk menghitung kuartil ketiga (75%) dari kolom tersebut.
* IQR = Q3 - Q1 digunakan untuk menghitung Interquartile Range (IQR), yaitu selisih antara Q3 dan Q1.
* lower = Q1 - 1.5 * IQR digunakan untuk menentukan batas bawah outlier berdasarkan metode IQR.
* upper = Q3 + 1.5 * IQR digunakan untuk menentukan batas atas outlier berdasarkan metode IQR.
* jumlah_outlier = df[(df[col] < lower) | (df[col] > upper)].shape[0] digunakan untuk menghitung jumlah data yang berada di luar batas bawah atau batas atas (outlier).
    * (df[col] < lower) | (df[col] > upper) adalah kondisi untuk memilih data yang termasuk outlier.
    * .shape[0] digunakan untuk mengambil jumlah baris (banyaknya outlier).
* outlier_summary.append([col, jumlah_outlier]) digunakan untuk menyimpan nama kolom dan jumlah outlier ke dalam list outlier_summary.
* outlier_table = pd.DataFrame(outlier_summary, columns=["Kolom", "Jumlah Outlier"]) digunakan untuk mengubah list ringkasan menjadi DataFrame dengan dua kolom, yaitu nama kolom dan jumlah outlier.
* outlier_table dipanggil untuk menampilkan tabel jumlah outlier pada setiap kolom numerik.

Berdasarkan analisis outlier menggunakan metode IQR pada dataset Iris, mayoritas data berada dalam rentang normal untuk setiap kolom numerik. Dari empat kolom numerik (sepal_length, sepal_width, petal_length, petal_width), hanya kolom sepal_width yang memiliki sejumlah kecil nilai outlier, sementara kolom lainnya relatif bersih tanpa outlier. Outlier pada sepal_width muncul di beberapa baris dan ditandai dalam tabel, namun jumlahnya sangat kecil dibandingkan total 150 baris, sehingga tidak signifikan untuk memengaruhi analisis keseluruhan. Secara keseluruhan, dataset Iris tergolong bersih dan stabil, sehingga sebagian besar model statistik atau machine learning dapat diterapkan tanpa perlu penanganan outlier yang besar.

```python
outlier_rows = pd.DataFrame()
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    temp = df[(df[col] < lower) | (df[col] > upper)]
    outlier_rows = pd.concat([outlier_rows, temp])
outlier_rows = outlier_rows.drop_duplicates()
outlier_rows
```

Penjelasan Kode:
* outlier_rows = pd.DataFrame() digunakan untuk membuat DataFrame kosong yang akan menyimpan seluruh baris data yang terdeteksi sebagai outlier.

* IQR = Q3 - Q1 digunakan untuk menghitung Interquartile Range (IQR), yaitu selisih antara Q3 dan Q1.

* upper = Q3 + 1.5 * IQR digunakan untuk menentukan batas atas outlier berdasarkan metode IQR.
* temp = df[(df[col] < lower) | (df[col] > upper)] digunakan untuk mengambil baris data yang memiliki nilai lebih kecil dari batas bawah atau lebih besar dari batas atas (termasuk outlier pada kolom tersebut).
* outlier_rows = pd.concat([outlier_rows, temp]) digunakan untuk menggabungkan (menambahkan) baris outlier yang ditemukan ke dalam DataFrame outlier_rows.
* outlier_rows = outlier_rows.drop_duplicates() digunakan untuk menghapus baris yang duplikat, karena satu baris bisa terdeteksi sebagai outlier di lebih dari satu kolom.

### B. Visualisasi Menghitung Jarak menggunakan Python

Analisis juga dilakukan menggunakan Python dengan bantuan Google Colab. Dataset yang digunakan adalah Iris Dataset.

```python
X = df[["sepal_length", "sepal_width", "petal_length", "petal_width"]].values

data_1 = X[0]
data_2 = X[1]

euclidean_distance = np.sqrt(np.sum((data_1 - data_2) ** 2))
manhattan_distance = np.sum(np.abs(data_1 - data_2))

print("Hasil Pehitungan jarak antar dua data :")
print(f"Euclidean Distance : {euclidean_distance:.4f}")
print(f"Manhattan Distance : {manhattan_distance:.4f}")
```

Penjelasan Kode:
* import numpy as np digunakan untuk mengimpor library NumPy untuk perhitungan numerik.
* X = df[["sepal_length", "sepal_width", "petal_length", "petal_width"]].values digunakan untuk mengambil fitur numerik dari Iris Dataset dan mengubahnya menjadi array.
* data_1 = X[0] mengambil data baris pertama.
* data_2 = X[1] mengambil data baris kedua untuk dibandingkan.
* np.sqrt(np.sum((data_1 - data_2) ** 2)) menghitung Euclidean Distance.
* np.sum(np.abs(data_1 - data_2)) menghitung Manhattan Distance.
* print() digunakan untuk menampilkan hasil perhitungan jarak kedua data.

Pada kode program tersebut diambil dua data pertama dari dataset kemudian dihitung jaraknya menggunakan metode Euclidean Distance dan Manhattan Distance. Perhitungan ini bertujuan untuk mengetahui tingkat kemiripan antara dua data berdasarkan empat fitur numerik yaitu sepal length, sepal width, petal length, dan petal width.
