## Menghitung Jarak Dataset Campuran
Pengukuran jarak antar data merupakan proses untuk menentukan tingkat kemiripan (similarity) atau perbedaan (dissimilarity) antara dua objek yang terdapat dalam suatu dataset. Nilai jarak digunakan untuk menunjukkan seberapa dekat atau jauh hubungan antara kedua objek tersebut.

Apabila nilai jarak sama dengan 0, maka kedua objek dianggap identik atau sama. Sebaliknya, semakin kecil nilai jarak, maka tingkat kemiripan antar objek semakin tinggi. Sementara itu, semakin besar nilai jarak, maka perbedaan antara objek akan semakin besar.

Metode pengukuran jarak yang digunakan biasanya disesuaikan dengan tipe data yang dianalisis. Untuk data nominal dan binary simetris, metode yang umum digunakan adalah Simple Matching. Pada data binary asimetris, digunakan koefisien Jaccard. Untuk data numerik, metode yang sering digunakan adalah Euclidean Distance atau Manhattan Distance yang termasuk dalam keluarga Minkowski Distance.

Pada data ordinal, nilai data terlebih dahulu diubah menjadi peringkat (ranking), kemudian dihitung menggunakan metode yang sama seperti data numerik. Sedangkan untuk data campuran, metode yang umum digunakan adalah Gower Distance.

Pengukuran jarak memiliki peran penting dalam berbagai teknik analisis data, seperti clustering, K-Nearest Neighbor (KNN), serta metode analisis data lainnya.

### Menghitung jarak antar data

```python
# Ambil 2 data pertama
data_1 = df.iloc[0]
data_2 = df.iloc[1]

# Kelompok kolom berdasarkan tipe data
kolom_numerik = ['age']
kolom_nominal = ['gender', 'course']
kolom_binary_asimetris = ['marital_status', 'depression', 'anxiety', 'panic_attack', 'specialist_treatment']
kolom_ordinal = ['year_of_study', 'cgpa']

# Definisi Ranking Data Ordinal
ranking_year = {
    'year 1': 1, 'year 2': 2, 'year 3': 3, 'year 4': 4
}

ranking_gpa = {
    '0 - 1.99': 1, '2.00 - 2.49': 2, '2.50 - 2.99': 3, 
    '3.00 - 3.49': 4, '3.50 - 4.00': 5
}

# Variabel Akumulator
jumlah_jarak = 0
jumlah_delta = 0

# NOMINAL
for kolom in kolom_nominal:
    if pd.notna(data_1[kolom]) and pd.notna(data_2[kolom]):
        jarak = 0 if str(data_1[kolom]).lower() == str(data_2[kolom]).lower() else 1
        jumlah_jarak += jarak
        jumlah_delta += 1

# BINARY ASIMETRIS
for kolom in kolom_binary_asimetris:
    v1 = str(data_1[kolom]).lower()
    v2 = str(data_2[kolom]).lower()

    if not (v1 == 'no' and v2 == 'no'):
        jarak = 0 if v1 == v2 else 1
        jumlah_jarak += jarak
        jumlah_delta += 1

# ORDINAL
for kolom in kolom_ordinal:
    val1 = str(data_1[kolom]).lower()
    val2 = str(data_2[kolom]).lower()
    if kolom == 'year_of_study':
        r1, r2, m = ranking_year.get(val1), ranking_year.get(val2), 4
    else:
        r1, r2, m = ranking_gpa.get(val1), ranking_gpa.get(val2), 5    
    if r1 is not None and r2 is not None:
        z1 = (r1 - 1) / (m - 1)
        z2 = (r2 - 1) / (m - 1)
        jumlah_jarak += abs(z1 - z2)
        jumlah_delta += 1

# NUMERIK
for kolom in kolom_numerik:
    rentang = df[kolom].max() - df[kolom].min()
    if rentang > 0:
        jarak = abs(data_1[kolom] - data_2[kolom]) / rentang
        jumlah_jarak += jarak
        jumlah_delta += 1

# HASIL
jarak_campuran = jumlah_jarak / jumlah_delta if jumlah_delta > 0 else 0

print(f"--- Hasil Perhitungan Jarak Gower ---")
print(f"Total Nilai Jarak (Pembilang) : {jumlah_jarak:.4f}")
print(f"Total Bobot/Delta (Penyebut)  : {jumlah_delta}")
print(f"Jarak Campuran Akhir          : {jarak_campuran:.4f}")
```

Penjelasan Kode:
* data_1 = df.iloc[0] dan data_2 = df.iloc[1] Digunakan untuk mengambil dua data pertama dari dataset yang akan dibandingkan jaraknya.
* kolom_numerik, kolom_nominal, kolom_binary_asimetris, kolom_ordinal Digunakan untuk mengelompokkan atribut berdasarkan tipe data agar perhitungan jarak dapat menggunakan metode yang sesuai.
* ranking_year dan ranking_gpa Digunakan untuk mengubah data ordinal menjadi nilai peringkat sehingga dapat dihitung jaraknya secara numerik.
* jumlah_jarak dan jumlah_delta Digunakan untuk menyimpan total jarak atribut dan jumlah atribut yang dihitung.
* Perulangan Nominal Menghitung jarak atribut nominal dengan nilai 0 jika sama dan 1 jika berbeda.
* Perulangan Binary Asimetris Menghitung jarak atribut biner dengan mengabaikan pasangan nilai 0–0.
* Perulangan Ordinal Mengubah peringkat menjadi nilai normalisasi lalu menghitung selisih absolut antar data.
* Perulangan Numerik Menghitung jarak numerik menggunakan normalisasi rentang nilai atribut.
* jarak_campuran = jumlah_jarak / jumlah_delta Digunakan untuk menghitung nilai jarak akhir antar data dengan metode Gower Distance.
* print(...) Digunakan untuk menampilkan total jarak, jumlah atribut yang dihitung, dan nilai jarak campuran akhir.
